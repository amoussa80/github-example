```python
quote_page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
import requests
```

```python
response = requests.get(quote_page)
```

```python
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
right_table=soup.find('table', class_='wikitable sortable')
```

```python
A=[]
B=[]
C=[]

for row in right_table.findAll("tr"):
    cells = row.findAll('td')
    if len(cells)==3: #Only extract table body not heading
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Postcode'])
df['Borough']=B
df['Neighbourhood']=C
df
```

```python
filtered_df = df[df.Borough != 'Not assigned']
filtered_df
```

```python
 for i in filtered_df['Neighbourhood']:
        if 'Not assigned' in i :
            filtered_df.loc[filtered_df['Neighbourhood'].str.contains('Not assigned'), 'Neighbourhood'] = filtered_df['Borough']

filtered_df
```

```python
grouped_df = filtered_df.groupby('Borough',as_index=False).agg(lambda Neighbourhood: ','.join(Neighbourhood))
grouped_df
#filtered_df = filtered_df.groupby('Borough')['Neighbourhood'].agg(lambda col: ','.join(col))
```

```python
filtered_df.to_csv('TorontoNeighbourhoods.csv')
```

```python
grouped_df.to_csv('TorontoNeighbourhoodsagg.csv')
```

```python
longlat_df = pd.read_csv('http://cocl.us/Geospatial_data')
longlat_df.head()
```

```python
longlat_df = longlat_df.rename(columns={'Postal Code': 'Postcode'})
```

```python
longlat_df.head()
```

```python
merged_df = pd.merge(filtered_df, longlat_df, on='Postcode', how='outer')
merged_df
```

```python
group_df = merged_df.groupby('Borough', as_index=False).aggregate({'Postcode': min,
                                                                  'Neighbourhood': ','.join,
                                                                  'Latitude': min,
                                                                  'Longitude':min})
group_df
```
